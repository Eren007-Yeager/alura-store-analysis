### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres = ["tienda 1", "tienda 2", "tienda 3", "tienda 4"]

for i, t in enumerate(tiendas):
    t["Total Pagado"] = t["Precio"] + t["Costo de envío"]
    facturacion = t["Total Pagado"].sum()
    print(f"Facturación total de {nombres[i]}: {facturacion}")
```

```python
import matplotlib.pyplot as plt

facturaciones = [t["Total Pagado"].sum() for t in tiendas]

plt.figure(figsize=(8,5))
plt.bar(nombres, facturaciones, color="skyblue")
plt.title("Facturación total por tienda")
plt.ylabel("Total en $")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Ventas por categoría

```python
print("Ventas por categoría:")
for i, t in enumerate(tiendas):
    print(f"\n{nombres[i]}:")
    print(t["Categoría del Producto"].value_counts())
```

```python
import seaborn as sns

for i, t in enumerate(tiendas):
    plt.figure(figsize=(8,5))
    categoria = t["Categoría del Producto"].value_counts()
    sns.barplot(x=categoria.values, y=categoria.index, palette="viridis")
    plt.title(f"Ventas por categoría - {nombres[i]}")
    plt.xlabel("Número de ventas")
    plt.ylabel("Categoría")
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
print("Calificación promedio:")
for i, t in enumerate(tiendas):
    promedio = round(t["Calificación"].mean(), 2)
    print(f"{nombres[i]}: {promedio}")
```

```python
promedios = [round(t["Calificación"].mean(), 2) for t in tiendas]

plt.figure(figsize=(8,5))
sns.barplot(x=nombres, y=promedios, palette="pastel")
plt.title("Calificación promedio por tienda")
plt.ylabel("Promedio de calificación")
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 4. Productos más y menos vendidos

```python
print("Productos más y menos vendidos:")
for i, t in enumerate(tiendas):
    ventas = t["Producto"].value_counts()
    mas_vendido = ventas.idxmax()
    menos_vendido = ventas.idxmin()
    print(f"\n{nombres[i]}:")
    print(f"  Producto más vendido: {mas_vendido} ({ventas.max()} ventas)")
    print(f"  Producto menos vendido: {menos_vendido} ({ventas.min()} ventas)")
```

```python
for i, t in enumerate(tiendas):
    ventas = t["Producto"].value_counts()
    top5 = ventas.head(5)
    bottom5 = ventas.tail(5)

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    sns.barplot(x=top5.values, y=top5.index, ax=axes[0], palette="Blues_d")
    axes[0].set_title(f"Más vendidos - {nombres[i]}")

    sns.barplot(x=bottom5.values, y=bottom5.index, ax=axes[1], palette="Reds_d")
    axes[1].set_title(f"Menos vendidos - {nombres[i]}")

    plt.tight_layout()
    plt.show()
```

# 5. Envío promedio por tienda

```python
print("Costo de envío promedio:")
for i, t in enumerate(tiendas):
    promedio_envio = round(t["Costo de envío"].mean(), 2)
    print(f"{nombres[i]}: {promedio_envio}")
```

```python
envio_promedio = [round(t["Costo de envío"].mean(), 2) for t in tiendas]

plt.figure(figsize=(8,5))
sns.barplot(x=nombres, y=envio_promedio, palette="coolwarm")
plt.title("Costo de envío promedio por tienda")
plt.ylabel("Costo en $")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
